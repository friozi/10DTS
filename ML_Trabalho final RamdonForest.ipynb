# Avaliação Comparativa dos Modelos

| Modelo                | Acurácia | Macro F1-Score | Melhor Classe | Ponto de Atenção |
|:----------------------|:---------|:--------------|:--------------|:----------------|
| Random Forest          | 0.64     | 0.61           | Standard (F1 = 0.69) | Desempenho limitado para Good (F1 = 0.52) |
| Árvore de Decisão      | 0.56     | 0.49           | Standard (F1 = 0.65) | Baixo desempenho para Good (F1 = 0.30) |
| Regressão Logística    | 0.55     | 0.39           | Standard (F1 = 0.68) | Dificuldade elevada na classe Good (F1 = 0.12) |


## Random Forest
O modelo Random Forest apresentou o melhor desempenho geral entre as alternativas testadas, atingindo uma acurácia de 64%. A classe "Standard" foi a mais bem classificada, com F1-score de 0.69, enquanto a classe "Good" apresentou um desempenho inferior (F1-score de 0.52). O macro F1-score de 0.61 indica um equilíbrio razoável entre as classes. Este modelo se mostra mais adequado para uso prático, com potencial de melhorias por meio de ajustes finos em seus parâmetros.



## Árvore de Decisão
O modelo de Árvore de Decisão obteve uma acurácia de 56%, inferior ao Random Forest. A classe "Standard" manteve um desempenho aceitável (F1-score de 0.65), mas a performance na classificação da classe "Good" foi bastante limitada (F1-score de 0.30). A árvore isolada demonstrou dificuldade em capturar padrões complexos dos dados, sugerindo que métodos de ensemble, como o Random Forest, são mais indicados.



## Regressão Logística
A Regressão Logística apresentou a menor acurácia entre os modelos (55%). Embora tenha mostrado um recall elevado para a classe "Standard" (86%), seu desempenho geral foi impactado pelo baixo resultado nas classes "Good" e "Poor", especialmente na classe "Good", onde o F1-score foi de apenas 0.12. O macro F1-score de 0.39 reforça o desbalanceamento de performance entre as categorias. Modelos lineares, como a Regressão Logística, tendem a apresentar limitações quando aplicados a bases de dados com alta cardinalidade de variáveis categóricas e relações não lineares.

# Random Forest


```python

# Importação de Bibliotecas
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Carregar uma amostra dos dados (50% dos dados)
train_df = pd.read_csv('train.csv', skiprows=lambda i: i > 0 and random.random() > 0.5)
test_df = pd.read_csv('test.csv', skiprows=lambda i: i > 0 and random.random() > 0.5)


print(train_df.head())
print(train_df.info())
print(train_df.describe())


plt.figure(figsize=(10, 6))
sns.countplot(x='Credit_Score', data=train_df)
plt.title('Distribuição da variável alvo (Credit Score)')
plt.show()


plt.figure(figsize=(12, 8))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

# Pré-processamento

```python


train_df.fillna(train_df.mean(numeric_only=True), inplace=True)
test_df.fillna(test_df.mean(numeric_only=True), inplace=True)

categorical_features = train_df.select_dtypes(include=['object']).columns


categorical_features = [col for col in categorical_features if col != 'Credit_Score']


train_df = pd.get_dummies(train_df, columns=categorical_features, drop_first=True, dtype='uint8')
test_df = pd.get_dummies(test_df, columns=categorical_features, drop_first=True, dtype='uint8')

print(train_df.shape)
print(test_df.shape)
print(train_df.memory_usage(deep=True).sum() / (1024 ** 2), "MB")

test_df = test_df.reindex(columns=train_df.drop('Credit_Score', axis=1).columns, fill_value=0)


X = train_df.drop('Credit_Score', axis=1)
y = train_df['Credit_Score']


top_columns = X.std().sort_values(ascending=False).head(300).index
X_reduced = X[top_columns]

```

# Treinando o modelo

```python

# Dividir os dados em treino e validação
X_train, X_val, y_train, y_val = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

# Treinamento do modelo Random Forest
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)


predictions = model.predict(X_val)

print(f'Acurácia Randon Forest: {accuracy_score(y_val, predictions):.2f}')
print('Relatório de Classificação:')
print(classification_report(y_val, predictions))


plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_val, predictions), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusão')
plt.xlabel('Predito')
plt.ylabel('Real')
plt.show()
```
